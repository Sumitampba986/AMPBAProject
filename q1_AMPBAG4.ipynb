Notes:
TfidfVectorizer package is used to Convert raw data string from the dataframe Software and Arts_Crafts_and_Sewing to a matrix of TF-IDF features. whereas CountVectorizer is used for occurence of each word in the document and TfidfTransformer It considers the importance of words not just within a document but also across the entire corpus.

```python
# Picking top and bottom 10k records and create a new data frame

# working df from software
df_software_h = base_df_software.head(10000)
df_software_t =   base_df_software.tail(10000)
#By considering both top and bottom 10k records creating new dataframe with 20k records as per question 
df_software = pd.concat([df_software_h,df_software_t])
print(df_software)
# working df from Arts_Crafts_and_Sewing
df_acs_h = base_df_Arts_Crafts_and_Sewing.head(10000)
df_acs_t =   base_df_Arts_Crafts_and_Sewing.tail(10000)
#By considering both top and bottom 10k records creating new dataframe with 20k records as per question 
df_acs = pd.concat([df_acs_h,df_acs_t])
print(df_acs)
```

```python
# Lower casing the software and acs review text's base dataframe dataset

df_software['reviewText'] = df_software['reviewText'].str.lower()
print(df_software)

df_acs['reviewText'] = df_acs['reviewText'].str.lower()
print(df_acs.head(1))
```

```python
# cleaning the dataset by removing punctuations in the dataframe dataset in SOFTWARE

df_sw_lower_punctuations = df_software

# This function will remove the puctuations from the input

def punc_cleaner(df_sw_lower_punctuations):
    if isinstance(df_sw_lower_punctuations,str):
        return re.sub(r'[^\w\s]', '', df_sw_lower_punctuations)
    else:
        df_sw_lower_punctuations
        
#Applying the above function on the dataframe column name "ReviewText"
df_sw_lower_punctuations['reviewText'] = df_sw_lower_punctuations['reviewText'].apply(punc_cleaner)
print(df_sw_lower_punctuations['reviewText'].head(1))




# cleaning the dataset by removing punctuations in the dataframe dataset in ARTS
df_acs_lower_punctuations = df_acs

# This function will remove the puctuations from the input
def punc_cleaner(df_acs_lower_punctuations):
    if isinstance(df_acs_lower_punctuations,str):
        return re.sub(r'[^\w\s]', '', df_acs_lower_punctuations)
    else:
        df_acs_lower_punctuations
        
#Applying the above function on the dataframe column name "ReviewText"
df_acs_lower_punctuations['reviewText'] = df_acs_lower_punctuations['reviewText'].apply(punc_cleaner)
print(df_acs_lower_punctuations['reviewText'].head(1))
```

```python
# Compute IDF of all words in these reviews for SOFTWARE reviews
#stop words (common words like "and," "the," etc.) are removed to focus on more meaningful terms.
idf = TfidfVectorizer(stop_words='english')


# Tokenize the review column for Software reviews
df_sw_lower_punctuations = pd.DataFrame(df_sw_lower_punctuations['reviewText'])


#Total 9 Records in ReviewText which has 0 Record which is dropped in below command as part of data cleaning.
df_sw_lower_punctuations =df_sw_lower_punctuations.dropna()


#applying the TfidfVectorizer (idf) to transform the 'reviewText' column in your DataFrame (df_sw_lower_punctuations).
df_sw_lower_punctuations_idf = idf.fit_transform(df_sw_lower_punctuations['reviewText'])

#get_feature_names_out() is used to get the feature names after applying the TfidfVectorizer. 
#It returns the terms (words) in the order they appear in the feature matrix.
df_sw_lower_punctuations_FN = idf.get_feature_names_out()


#creating a DataFrame df_sw_lower_punctuations_FN_idf that contains the words and their corresponding inverse document frequency (IDF) values.
#This can be valuable information for understanding the importance of each term in your text data.
df_sw_lower_punctuations_FN_idf = pd.DataFrame({'Word':df_sw_lower_punctuations_FN,'IDF':idf.idf_})
head = df_sw_lower_punctuations_FN_idf.sort_values(by=['IDF'], ascending=False).head(30)
tail = df_sw_lower_punctuations_FN_idf.sort_values(by=['IDF'], ascending=False).tail(30)

df_sw_lower_punctuations_FN_idf = pd.concat([head,tail])
print(df_sw_lower_punctuations_FN_idf)



# Compute IDF of all words in these reviews for Arts_Crafts_and_Sewing
idf = TfidfVectorizer(stop_words='english')


# Tokenize the review column for Software reviews
df_acs_lower_punctuations = pd.DataFrame(df_acs_lower_punctuations['reviewText'])
df_acs_lower_punctuations =df_acs_lower_punctuations.dropna()

#Computing the IDF of All words from the column name "ReviewText"
df_acs_lower_punctuations_idf = idf.fit_transform(df_acs_lower_punctuations['reviewText'])
df_acs_lower_punctuations_FN = idf.get_feature_names_out()
#Creating separate column word and Respective IDF values as Output.
df_acs_lower_punctuations_FN_idf = pd.DataFrame({'Word':df_acs_lower_punctuations_FN,'IDF':idf.idf_})
head = df_acs_lower_punctuations_FN_idf.sort_values(by=['IDF'], ascending=False).head(30)
tail = df_acs_lower_punctuations_FN_idf.sort_values(by=['IDF'], ascending=False).tail(30)

df_acs_lower_punctuations_FN_idf = pd.concat([head,tail])
print(df_acs_lower_punctuations_FN_idf)
```
