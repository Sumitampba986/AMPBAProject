```python
# READING THE FILE and creating base dataframe for Software and Arts_Crafts_and_Sewing revoew comments

# PART1
# Source file path for software reviews
#json_file_path = '/Users/anirudhyadav/Documents/PlayArea/EDA/Software.json'

#Testing Path
json_file_path = r'C:\Users\sumit\Desktop\ISB\Classwork\Text Analytics_R2\Grp Assign\Software.json'

# Read the JSON file and load into the dataframe
df_sowftware = pd.read_json(json_file_path, lines=True,nrows=10)

# Creating the base dataframe 
base_df_software = df_sowftware[['reviewerID','reviewText']]


# *****************************************************#
# Part2
# Source file path for Arts_Crafts_and_Sewing 
#json_file_path = '/Users/anirudhyadav/Documents/PlayArea/EDA/Arts_Crafts_and_Sewing.json'

#Testing Path 
json_file_path = r'C:\Users\sumit\Desktop\ISB\Classwork\Text Analytics_R2\Grp Assign\Arts_Crafts_and_Sewing_1.json'

# Read the JSON file and load into the dataframe
df_Arts_Crafts_and_Sewing = pd.read_json(json_file_path, lines=True,nrows=10)

# Creating the base dataframe 
base_df_Arts_Crafts_and_Sewing = df_Arts_Crafts_and_Sewing[['reviewerID','reviewText']]


# *****************************************************#
# Creating single dataframe for both the categories

combined_df = pd.concat([base_df_software,base_df_Arts_Crafts_and_Sewing], ignore_index=True)
print(combined_df)
```

```python
# Download the punkt tokenizer (only needed once)
nltk.download('punkt')
```

```python
# Tokenizing the combined data frame

#word_tokenize this would convert reviewText dataframe to root words 
combined_df['Tokenized_Review'] = combined_df['reviewText'].apply(word_tokenize)
# Print the DataFrame with tokenized reviews
print(combined_df)
```

```python
# Lemmatization of the combined dataframe
combined_df_lm = combined_df
# Lemmatization function
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens

# Apply lemmatization to the 'Tokenized_Review' column
combined_df_lm['Lemmatized_Review'] = combined_df_lm['Tokenized_Review'].apply(lemmatize_tokens)

# Print the DataFrame with lemmatized reviews
print(combined_df_lm)
```

```python
#  Part of speach tagging
combined_df_lm_pos = combined_df_lm

def pos_tagging(tokens):
    pos_tags = pos_tag(tokens)
    return pos_tags

# Apply part-of-speech tagging to the 'Lemmatized_Review' column
combined_df_lm_pos['POS_Tags_Review'] = combined_df_lm_pos['Lemmatized_Review'].apply(pos_tagging)

# Print the DataFrame with part-of-speech tags
print(combined_df_lm_pos)
```
